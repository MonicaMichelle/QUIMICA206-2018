```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
wine=datasets.load_wine()
wine
```

```python
wine_dataframe=pd.DataFrame(data=wine.data,columns=wine.feature_names)
wine_dataframe["target"]=wine.target
wine_dataframe
```

```python
x=wine_dataframe["ash"]
print(x.mean())
y=wine_dataframe["flavanoids"]
col=(wine_dataframe["target"]+1)*2
print(y.mean())
ash=wine_dataframe["ash"]*100
flavanoids=wine_dataframe["flavanoids"]
print(x.std())
print(y.std())
```

```python
plt.scatter(x,y,c=col,s=ash,alpha=0.5)
plt.xlabel("cantidad de cenizas")
plt.ylabel("cantidad de flavanoides")
plt.title("cenizas vs flavanoides")
plt.grid(True)
plt.show()
```
